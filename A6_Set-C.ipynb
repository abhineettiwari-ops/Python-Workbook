```python
def sum_of_digits(n):              #Function is created to find sum of digits using recursion
    if n == 0:                     # This is the base condition to stop recursion
        return 0                   # If number becomes 0, return 0
    else:
        return (n % 10) + sum_of_digits(n // 10)
                                   #n%10 gives last digit
                                   #n//10 removes last digit and calls function again
num = int(input("Enter a number: "))    # Take number from user
print("Sum of digits:", sum_of_digits(num))  # Call function and print result
```

<h4>2) Write a function that accepts a list of numbers and returns a dictionary with count of even 
<h4>and odd numbers. 
<h5>Example: [1,2,3,4,5] should return {'even': 2, 'odd': 3} 

```python
def count_even_odd(lst):                 # Function is created to count even and odd numbers
    result = {"even":0, "odd":0}         # Dictionary is used to store counts

    for n in lst:                        # Loop goes through each number in the list
        if n % 2 == 0:                   # Check if number is even
            result["even"] = result["even"] + 1   # Increase even count
        else:
            result["odd"] = result["odd"] + 1     # Increase odd count

    return result                        # Return dictionary with counts
numbers = [1,2,3,4,5]                    # Given list
print(count_even_odd(numbers))           # Call function and print result
```
