<h4>1) Write a program to find the union, intersection, and difference of two sets entered by the 
user.

```python
s1=set(input("Enter first set elements: ").split())   
s2=set(input("Enter second set elements: ").split())  
union_set=s1|s2           #this line joins both sets and keeps only unique elements. 
intersection_set=s1&s2    #This line finds elements that are present in both sets 
difference_set=s1-s2      #This line keeps only those elements that are present in the first set and not present in the second set.
print("Union:",union_set)                 
print("Intersection:",intersection_set)   
print("Difference:",difference_set)       
```

<h4>2) Write a program to check if a set is a subset of another set.

```python
s1=set(input("Enter first set elements: ").split())   
s2=set(input("Enter second set elements: ").split())  

if s1.issubset(s2):           # Check whether all elements of the first set are present in the second set.
    print("First set is a subset of second set")
else:
    print("First set is NOT a subset of second set")
```

<h4>3) Write a program to remove duplicates from a list using sets.

```python
lst=[1,2,2,3,4,4,5,5]       
s=set(lst)                   #Convert list to set (duplicates removed)
lst=list(s)                  # Convert set back to list
print("List without duplicates:",lst)   
```

<h4>4) Write a program to find elements that are present in the first set but not in the second set. 

```python
s1={1,2,3,4}                 # First set
s2={3,4,5}                   # Second set
result=s1-s2                 # Elements in s1 but not in s2
print("Result:",result)      
```

<h4>5) Write a program to check if two sets are disjoint (have no common elements).

```python
s1=set(input("Enter first set elements: ").split())    #split() is used to separate input values by space
s2=set(input("Enter second set elements: ").split())    #set() is used to store only unique elements
if s1.isdisjoint(s2):        #isdijoint() checks if both sets have NO common elements
    print("Sets are disjoint")
else:
    print("Sets are NOT disjoint")
```
