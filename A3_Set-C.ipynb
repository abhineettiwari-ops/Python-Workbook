<h4>1) Write a program to find common elements between two tuples. 

```python
t1=(1,2,3,4,5)              
t2=(3,4,5,6,7)              
common=()                   #Empty tupple to store common elements

for i in range(len(t1)):    
    if t1[i] in t2:         #Check if element is present in second tuple
        common=common+(t1[i],)  #This statement adds a common element to the tuple by concatenating a single-element tuple.

print("Common elements:",common) 
```

<h4>2) Write a program to create a tuple of squares of numbers from 1 to n. 
<h4>Example: If n=5, output should be (1, 4, 9, 16, 25)

```python
n=int(input("Enter a number: "))   
t=()                               #Empty tuple to store squares

for i in range(1,n+1):             #Loop from 1 to n
    t=t+(i*i,)                     # Add square of number to tuple

print("Tuple of squares:",t)      
```
