<h4>1) Write a program to create a dictionary with roll number as key and name as value for 5 
students. Display all key-value pairs.

```python
students={}                             # Create empty dictionary to store student data

for i in range(5):                      # Loop is used to take details of 5 students
    roll=int(input("Enter roll number: "))   # Roll number is taken as key because it is unique
    name=input("Enter student name: ")       # Name is taken as value for the roll number
    students[roll]=name                     # Store roll-number and name as key-value pair

print("Student Details:")               # Print heading before displaying data
for key,value in students.items():      # items() is used to get both key and value together
    print(key,":",value)                # Print roll number and student name
```

<h4>2) Write a program to count the frequency of each word in a given sentence using a 
dictionary. 
<h5>Example: "hello world hello" should give {'hello': 2, 'world': 1}

```python
sentence=input("Enter a sentence: ")        #Take sentence from user

words=sentence.split()                      #split() is used to separate sentence into words

freq={}                                     #Empty dictionary to store word counts

for word in words:                          # Loop through each word in the list
    if word in freq:                        # Check if word already exists in dictionary
        freq[word]=freq[word]+1             # Increase count if word is repeated
    else:
        freq[word]=1                        # Set count as 1 for first occurrence

print(freq)                                 # Print dictionary with word frequencies
```

<h4>3) Write a program to merge two dictionaries entered by the user.

```python
d1={}                                      
d2={}                                      

n1=int(input("Enter number of items in first dictionary: "))   # How many key-value pairs in d1

for i in range(n1):                        #Loop to enter items for first dictionary
    key=input("Enter key: ")               # Take key from user
    value=input("Enter value: ")           # Take value for the key
    d1[key]=value                          # Store key-value pair in first dictionary
n2=int(input("Enter number of items in second dictionary: "))  # How many key-value pairs in d2
for i in range(n2):                       #Loop to enter items for second dictionary
    key=input("Enter key: ")               # Take key from user
    value=input("Enter value: ")           #Take value for the key
    d2[key]=value                          # Store key-value pair in second dictionary

d1.update(d2)                              # update() is used to add all items of d2 into d1

print("Merged Dictionary:",d1)             # Print the merged dictionary
```

<h4>4) Write a program to find the key with the maximum value in a dictionary. 

```python
marks={"Math":85,"Science":90,"English":88}   # Dictionary with subjects as keys and marks as values
max_key=list(marks.keys())[0]                # Assume first key has the maximum value
for key in marks:                             # Loop through each key in dictionary
    if marks[key] > marks[max_key]:           # Compare values of keys
        max_key=key                           # Update key if higher value is found

print("Key with maximum value:",max_key)      
print("Maximum value:",marks[max_key])       
```

<h4>5) Write a program to check if a given key exists in a dictionary.

```python
student={"name":"Rahul","age":20,"grade":"A"}   # Dictionary with student details
key=input("Enter key to search: ")              # Take key from user
if key in student:                              # in is used to check key existence
    print("Key exists in the dictionary")
else:
    print("Key does NOT exist in the dictionary")
```
