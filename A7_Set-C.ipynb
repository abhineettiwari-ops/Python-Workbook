<h4>1) Create an abstract class Shape with abstract method area(). Create subclasses Rectangle, 
Circle, and Triangle that implement the area method.

```python
from abc import ABC, abstractmethod          # ABC and abstractmethod are used for abstraction

class Shape(ABC):                            # Shape is an abstract class

    @abstractmethod
    def area(self):                          # Abstract method to calculate area
        pass                                 # pass is used because method has no body
class Rectangle(Shape):                     # Rectangle class inherits Shape
    def __init__(self, l, w):
        self.l = l                           # store length
        self.w = w                           # store width

    def area(self):
        return self.l * self.w               # area formula for rectangle
class Circle(Shape):                        # Circle class inherits Shape
    def __init__(self, r):
        self.r = r                           # store radius

    def area(self):
        return 3.14 * self.r * self.r        # area formula for circle
class Triangle(Shape):                      # Triangle class inherits Shape
    def __init__(self, b, h):
        self.b = b                           # store base
        self.h = h                           # store height

    def area(self):
        return 0.5 * self.b * self.h         # area formula for triangle
rect = Rectangle(5, 4)                      # object of Rectangle
cir = Circle(3)                             # object of Circle
tri = Triangle(6, 2)                        # object of Triangle

print("Rectangle Area:", rect.area())       # calls rectangle area
print("Circle Area:", cir.area())           # calls circle area
print("Triangle Area:", tri.area())         # calls triangle area
```

<h4>2) Create a class Library that manages a collection of books. Implement methods to add 
books, remove books, search books, and display all books.

```python
class Library:                                  # Class is used to group all book-related work

    def __init__(self):                         # Constructor runs automatically when object is created
        self.books = []                         # List is used because it can store many books

    def add_book(self, book):                   # Method to add a new book
        self.books.append(book)                # append() is used to add book to the list

    def remove_book(self, book):                # Method to remove a book
        if book in self.books:                  # Check is needed to avoid error
            self.books.remove(book)             # remove() deletes the book from list
        else:
            print("Book not found")             # Message shown if book is not present

    def search_book(self, book):                # Method to search a book
        if book in self.books:                  # in keyword checks availability
            print("Book available")             # Printed if book exists
        else:
            print("Book not available")         # Printed if book does not exist

    def display_books(self):                    # Method to show all books
        print(self.books)                       # Printing list shows all stored books
```

```python
lib = Library()                                 # Object is created to use the class

lib.add_book("Python")                          # Add first book
lib.add_book("Data Science")                    # Add second book

lib.display_books()                             # Show all books

lib.search_book("Python")                       # Search a book
lib.remove_book("Python")                       # Remove a book

lib.display_books()                             # Show books again
```
