<h4>2) Write a program to sort a dictionary by its values in descending order.

```python
marks={"Math":85,"Science":90,"English":88}   

items=list(marks.items())                    # Convert dictionary to list
for i in range(len(items)):                 
    for j in range(i+1,len(items)):
        if items[i][1]<items[j][1]:          # Compare values
            items[i],items[j]=items[j],items[i]  # Swap items
marks=dict(items)                            # Convert list back to dictionary

print(marks)                                
```
