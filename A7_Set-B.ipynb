<h4>1) Create a class hierarchy: Animal -> Mammal -> Dog. Demonstrate multilevel inheritance. 

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")
        # this method shows a general behavior of all animals
class Mammal(Animal):
    def walk(self):
        print("Mammal can walk")
        # this method shows a behavior common to mammals
class Dog(Mammal):
    def bark(self):
        print("Dog barks")
        # this method shows a specific behavior of dog
d = Dog()          # object of Dog class is created

d.speak()          # Dog uses method of Animal class
d.walk()           # Dog uses method of Mammal class
d.bark()           # Dog uses its own method
```

<h4>2) Create a class Employee with class method to keep count of total employees. 

```python
class Employee:
    count = 0
    def __init__(self):
        Employee.count += 1
    @classmethod
    def total(cls):
        return cls.count

e1 = Employee()
e2 = Employee()
e3 = Employee()

print(Employee.total())
```

<h4>3) Create a class Book with special methods __str__, __len__, and __eq__.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title          # title is stored to identify the book
        self.pages = pages          # pages are stored to know book size

    def __str__(self):
        return self.title           
        # this method is used so that book object prints its name instead of memory address

    def __len__(self):
        return self.pages           
        # this method is used so len(book) returns number of pages

    def __eq__(self, other):
        return self.pages == other.pages  
        # this method is used to compare two books using == operator
```

```python
b1 = Book("Python Basics", 200)     # first book object is created
b2 = Book("Advanced Python", 200)   # second book object is created

print(b1)                           # __str__ is called automatically
print(len(b1))                      # __len__ is called automatically
print(b1 == b2)                     # __eq__ is called automatically
```

<h4>4) Create a class Temperature with property decorators for Celsius and Fahrenheit 
conversion.

```python
class Temperature:                         # Class is created to handle temperature values

    def __init__(self, celsius):           # Constructor is used to set initial value
        self._celsius = celsius            # _celsius stores temperature safely inside object

    @property
    def celsius(self):                     # Property is used to access value like variable
        return self._celsius               # Returns stored celsius value

    @celsius.setter
    def celsius(self, value):              # Setter is used to update celsius value
        self._celsius = value              # Updates the internal temperature value

    @property
    def fahrenheit(self):                  # Property is used to get fahrenheit value
        return (self._celsius * 9/5) + 32  # Formula converts celsius to fahrenheit
t = Temperature(25)                        # Object is created with 25 degree celsius

print("Celsius:", t.celsius)              # Prints celsius value using property
print("Fahrenheit:", t.fahrenheit)        # Prints fahrenheit after conversion

t.celsius = 30                             # Updates celsius value using setter
print("Updated Fahrenheit:", t.fahrenheit)# Prints updated fahrenheit value
```
