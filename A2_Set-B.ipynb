<h3>1) Write a program to find the first non-repeating character in a string.

```python
s=input("Enter a string: ")        
for i in range(len(s)):            # This loop checks each position of the string one by one.
    if s.count(s[i])==1:           # Check if character appears only one time 
        print("First non-repeating character:",s[i])  
        break                      # it will help to stop 
```

<h3>2) Write a program to check if two strings are anagrams (contain same characters in different 
order). 
<h4>Example: "listen" and "silent" are anagrams.

```python
s1=input("Enter first string: ")      
s2=input("Enter second string: ")     

flag=True                             # Assume strings are anagrams

if len(s1)!=len(s2):                  # This condition checks whether the two strings have different lengths, in which case they cannot be anagrams.
    flag=False                        # They cannot be anagrams
else:
    for i in range(len(s1)):          # This loop checks each letter of the first string one by one.
        if s1.count(s1[i])!=s2.count(s1[i]):  # Compare character count
            flag=False                # Not anagrams
            break

if flag:
    print("Anagrams")                
else:
    print("Not anagrams")             
```

<h3>3) Write a program to convert a string to title case without using the title() method. 
<h4>Example: "python programming" should become "Python Programming"

```python
text=input("Enter a string: ")        
words=text.split()       # Split string into words             
result=""                            

for word in words:                   
    first=word[0].upper()             # Convert first letter to uppercase
    rest=word[1:]                    # Get remaining letters of the word
    result=result+first+rest+" "      #used to join word with space

print("Title case string:",result)    
```

<h3>4) Write a program to replace all occurrences of a substring with another substring in a given 
string. 

```python
text=input("Enter a string: ")            # This is the main sentence/string
old=input("Enter substring to replace: ") #here the user will enter what he want to replace 
new=input("Enter new substring: ")        # here the user will enter new word which he want at  the place of old

result=text.replace(old,new)              # Replaces all old words with new words

print("Updated string:",result)           
```
