<h3>1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'.

```python
n = int(input("Enter a number: "))   #Taking input from the user 
for num in range(2, n + 1):          #here we are starting our loop from 2 because the smallest prime number is 2 
    is_prime = True                  
    for i in range(2, num):
        if num % i == 0:                 #If divisible then it is not a prime number
            is_prime = False
            break
    if is_prime:                         #if the number is prime after division then print 
        print(num)   
```

<h3>2) Write a program that accept two integer values and if both are equal then prints "SAME 
identity" otherwise prints "DIFFERENT identity".

```python
num1 = int(input("enter the first number:"))      #Taking input 
num2 = int(input("enter the second number:")) 
#Comparison between num1 and num2
if num1==num2:      #Using Relational Operator to check the identity
    print("SAME identity")
else:
    print("DIFFERENT identity")
    
```

<h3>3) Write a program to display following pattern: 
<h4>1 2 3 4 
<h4>1 2 3 
<h4>1 2 
<h4>1

```python
# This outer loop decides how many rows will be printed
# It starts from 4 and goes down to 1
for i in range(4, 0, -1):          #-1 is used to decrease the value by 1 in each loop so that rows print from 4 to 1.
    for j in range(1, i + 1):      #This loop prints numbers starting from 1 and ending at the current row number and +1 is used so that the loop includes the number i while printing.
        print(j, end=" ")          # end=" " is used so that the next number prints on the same line
    print()                        #It tells the python to go to the next line and start printing there
```

<h3>4) Write a program to reverse a given number.

```python
num = int(input("Enter a number: "))
rev = 0    # This variable is used to store the reversed number 
while num > 0:     #Using while loop to run the code as long as the number is greater thean 0
    digit = num % 10    # using modulus operator to get the last digit of the number 
    rev = rev * 10 + digit           #if rev==4 and digit == 5 then rev= 4*10+5=45
    num = num // 10                  #Using  Floor division to remove the last digit
print("Reversed number is:", rev)
```
