<h4>1) Write a program to find symmetric difference between two sets.

```python
s1={1,2,3,4}                 # First set created to store unique elements
s2={3,4,5,6}                 # Second set created for comparison
sym_diff=s1^s2               # ^ is used to get elements not common in both sets
print("Symmetric Difference:",sym_diff)  
```

<h4>2) Write a program to add multiple elements to a set at once from user input.

```python
s={1,2,3}                                      # Initial set with some elements

new_elements=input("Enter elements to add: ") # Take multiple elements from user as a string
s.update(new_elements.split())                # split() separates input values and update() adds all elements to the set at onc
print("Updated set:",s)                       # Print the updated set
```

<h4>3) Write a program to find the length of a set and check if a specific element exists in it.

```python
s={1,2,3,4,5}                

print("Length of set:",len(s)) # len() is used to count elements in the set

element=int(input("Enter element to search: ")) # Convert input to integer

if element in s:              # in checks if element exists in the set
    print("Element is present in the set")
else:
    print("Element is NOT present in the set")
```
