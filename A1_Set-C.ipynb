<h3>1) Write a Sequential search function which searches an item in a sorted list. The function 
should return the index of element to be searched in the list.

```python
def sequential_search(arr,key):        # Function to search an element in the list
    for i in range(len(arr)):          # Loop through the list index by index
        if arr[i]==key:                # Check if current element matches key
            return i                   # Return index if element is found
    return -1                          # Return -1 if element is not found
numbers=[10,20,30,40,50]               # Sorted list of numbers
key=int(input("Enter element to search: "))  # Take element to search from user
index=sequential_search(numbers,key)  # Call the search function
if index!=-1:                          # Check if element is found
    print("Element found at index:",index)  # Print index of found element
else:
    print("Element not found")         # Print message if element not found
```
