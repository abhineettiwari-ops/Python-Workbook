<h4>1) Write a program that takes two sentences as input and finds: 
<h5>● Common words in both sentences 
<h5>● Words unique to the first sentence 
<h5>● Words unique to the second sentence

```python
s1=input("Enter first sentence: ")       
s2=input("Enter second sentence: ")       
set1=set(s1.split())                      # split() breaks sentence into words, set removes duplicates
set2=set(s2.split())                     # convert second sentence into a set of words

common=set1 & set2                        # & is used to find common words
only_first=set1 - set2                    #words present only in first sentencw
only_second=set2 - set1                   # words present only in second sentence
print("Common words:",common)            
print("Unique to first sentence:",only_first)   
print("Unique to second sentence:",only_second) 
```

<h4>2) Write a program to perform all set operations (union, intersection, difference, symmetric 
difference) on three sets.

```python
s1={1,2,3}                   
s2={3,4,5}                    
s3={5,6,7}                    

union_set=s1|s2|s3          # | is used to combine all elements from three sets
                              # duplicates are automatically removed   
intersection_set=s1&s2&s3     # & is used to find elements common in all three sets
difference_set=s1-s2-s3       # - is used to remove elements of other sets from first set
sym_diff=(s1^s2)^s3            # ^ is used to find elements not common between sets and it is applied twice to include third set
                              
print("Union:",union_set)                
print("Intersection:",intersection_set)   
print("Difference:",difference_set)      
print("Symmetric Difference:",sym_diff)   
```
