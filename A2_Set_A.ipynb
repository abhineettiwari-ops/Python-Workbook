<h3>1) Write a program to count the number of vowels in a given string. 

```python
s=input("Enter a string: ")       
vowels=0                           # Variable to store count of vowels

for i in s:                       # Loop through each character in the string
    if i in "aeiouAEIOU":          # Check if the character is a vowel
        vowels=vowels+1            # Increase vowel count by 1

print("Number of vowels:",vowels)  # Print total number of vowels

```

<h3>2) Write a program to check if a string is a palindrome (reads the same forwards and 
backwards).

```python
s=input("Enter a string: ")        
rev=""                             # It is to string to store reversed string

for letter in s:                   #Using for Loop
    rev=letter+rev                 # This line builds the reversed string by adding each new letter to the front

if s==rev:                         # Compare original string with reversed string
    print("Palindrome") 
else:
    print("Not a palindrome")      
    
```

<h3>3) Write a program to reverse a string without using built-in reverse functions.

```python
s=input("Enter a string: ")       
rev=""                             # Empty string to store reversed string

for letter in s:                   # Loop through each character in the string
    rev=letter+rev                 # Add character in front to reverse the string

print("Reversed string:",rev)      
```

<h3>4) Write a program to count the frequency of each character in a string. 
<h3>Example: For "hello", output should be h:1, e:1, l:2, o:1 

```python
s=input("Enter a string: ")        
freq={}                            # Using dictionary to store character counts

for ch in s:                       # Loop through each character in the string
    if ch in freq:                 # Check if character already exists in dictionary
        freq[ch]=freq[ch]+1        # Increase count by 1
    else:
        freq[ch]=1                 # Add character with count 1
for ch in freq:                    # Loop through dictionary keys
    print(ch,":",freq[ch])         
```

<h3>5) Write a program to remove all whitespace from a string. 

```python
text=input("Enter a string: ")        
result=""                             # Empty string to store final output
for s in text:                        
    if s!=" ":                       # Check if character is not a space  by using !=
        result=result+s              # Add character to result string
print("String without spaces:",result)  
```
