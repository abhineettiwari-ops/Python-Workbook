<h4>1) Write a function to find the maximum of three numbers.

```python
def find_max(a,b,c):              # Functionm is created to find the largest of three numbers
    max_num=a                     # First number is taken as maximum to start comparison
    if b>max_num:                 # Compare second number with current maximum
        max_num=b                 # Update maximum if second number is larger
    if c>max_num:                 # Compare third number with current maximum
        max_num=c                 # Update maximum if third number is larger

    return max_num                # Return the final maximum value

```

```python
result=find_max(10,25,15)         # Calling 
print("Maximum number is:",result) 
```

<h4>2) Write a function to check if a number is prime or not.

```python
def is_prime(n):                     # Function is created to check whether a number is prime
    if n <= 1:                       # Numbers 1 or less cannot be prime by definition
        return False                 # Return False because prime numbers start from 2
    for i in range(2,n):             # Loop checks divisibility from 2 up to n-1
        if n % i == 0:               # Modulus operator checks if n divides evenly
            return False             # If divisible, number is not prime
    return True                      # If no divisor is found, number is prime
num=int(input("Enter a number: "))   # Input is taken so user can test any number

if is_prime(num):                    # Function is called to check prime or not
    print("Number is Prime")         # Printed when function returns True
else:
    print("Number is Not Prime")     # Printed when function returns False
```

<h4>3) Write a function to calculate the factorial of a number (both iterative and recursive).

<h5>Iterative Method (using loop)

```python
def factorial_iterative(n):        # Function is created to find factorial using loop
    fact = 1                       # Variable is used to store factorial result
    for i in range(1, n+1):        # Loop runs from 1 to n
        fact = fact * i            # Multiply each number to get factorial
    return fact                    # Return final factorial value
num = int(input("Enter a number: "))    # Take number from user
print("Factorial (Iterative):", factorial_iterative(num))  # Call function and print result
```

<h4>4) Write a function that accepts a list and returns a new list with unique elements.

```python
def unique_list(lst):        
    unique=[]               
    for x in lst:            
        if x not in unique:  
            unique.append(x)
    return unique
numbers=[1,2,3,2,4,1,5]      
print(unique_list(numbers)) 
```

<h4>5) Write a function to calculate the sum of all even numbers in a list. 

```python
def sum_even(lst):
    total=0
    for x in lst:
        if x%2==0:
            total=total+x
    return total
numbers=[1,2,3,4,5,6,8]
print(sum_even(numbers))
```
