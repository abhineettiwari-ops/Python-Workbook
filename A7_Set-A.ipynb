<h4>1) Create a class called Rectangle with attributes length and width. Add methods to 
calculate area and perimeter. 

```python
class Rectangle:                      #we create class to represent a rectangle 
    def __init__(self, length, width):  # __init__ is used to set values when object is created
        self.length = length           # Store length inside the object
        self.width = width            # Store width inside the objec
    def area(self):                   # Method is created to find area
        return self.length * self.width   # Area is length multiplied by width
    def perimeter(self):              # Method to calculate perimeter
        return 2 * (self.length + self.width)  # Perimeter formula
```

```python
r = Rectangle(8, 4)               # Object is created with length and width
print("Area:", r.area())          # Area method is called
print("Perimeter:", r.perimeter())# Perimeter method is called
```

<h4>2) Create a class BankAccount with methods to deposit, withdraw, and display balance. 

```python
class BankAccount:                   # Class is created to represent a bank account
    def __init__(self, balance):     # __init__ is used to set initial balance
        self.balance = balance       # Balance is stored inside the object
    def deposit(self, amount):       # Method is used to add money
        self.balance = self.balance + amount   # Add amount to balance
    def withdraw(self, amount):      # Method is used to remove money
        if amount <= self.balance:   # Check if enough balance is available
            self.balance = self.balance - amount
        else:
            print("Insufficient balance")  # Message if balance is less
    def show_balance(self):          # Method is used to display balance
        print("Balance:", self.balance)

acc = BankAccount(1000)              # Object is created with initial balance

acc.deposit(500)                     # Deposit money
acc.withdraw(300)                    # Withdraw money
acc.show_balance()                   # Display current balance
```

<h4>3) Create a parent class Vehicle and child classes Car and Bike that inherit from it. Override 
a method in child classes.

```python
class Vehicle:
    def move(self):
        print("Vehicle is moving")
class Car(Vehicle):
    def move(self):
        print("Car is moving on road")
class Bike(Vehicle):
    def move(self):
        print("Bike is moving on road")
v = Vehicle()
c = Car()
b = Bike()

v.move()
c.move()
b.move()
```

<h4>4) Create a class Student with private attributes and public methods to access them.

```python
class Student:
    def __init__(self, name, marks):
        self.__name = name      # __name is made private to protect student name
        self.__marks = marks    # __marks is made private to protect marks
    def get_name(self):
        return self.__name      # this method is used to safely access private name
    def get_marks(self):
        return self.__marks     # this method is used to safely access private marks
s = Student("Rahul", 85)        # object is created to store student details
print("Name:", s.get_name())   # private name is accessed using public method
print("Marks:", s.get_marks()) # private marks are accessed using public method
```

<h4>5) Create a class Circle with method to calculate area. Overload the + operator to add two 
circles (add their radii).

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius        # radius is stored inside object to use in calculations
    def area(self):
        return 3.14 * self.radius * self.radius  
        # formula is used to calculate area of circle
    def __add__(self, other):
        return Circle(self.radius + other.radius)
        # + operator is overloaded to add radii of two circles
c1 = Circle(5)                      # first circle object
c2 = Circle(3)                      # second circle object

c3 = c1 + c2                        # + operator calls __add__ method

print("Area of first circle:", c1.area())
print("Area of second circle:", c2.area())
print("Radius after adding circles:", c3.radius)
```
