<h4>1) Write a program to find all keys with the same value in a dictionary.

```python
data={"a":1,"b":2,"c":1,"d":3,"e":2}   # Dictionary is created to store keys and values
value=input("Enter value to search: ") # Input is taken to know which value we want to match
keys_list=[]                           # Empty list is used to store matching keys
for key in data:                       # Loop is used to check each key in the dictionary
    if str(data[key])==value:          # Value is converted to string to match user input
        keys_list.append(key)          # Matching key is added to the list
print("Keys with the same value:",keys_list)  # Print all keys having the given value
```

<h4>2) Write a program to create a dictionary where keys are numbers from 1 to n and values are 
their cubes.

```python
n = int(input("Enter a number: "))
cube_dict = {}
for i in range(1, n + 1):
    cube_dict[i] = i ** 3
print(cube_dict)
```
