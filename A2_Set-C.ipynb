<h3>1) Write a program to find the longest word in a sentence.

```python
sentence=input("Enter a sentence: ")    
words=sentence.split()                   # Split sentence into list of words
longest=words[0]                         # This line stores the first word as the longest word initially for comparison.
for i in range(len(words)):              #This loop is used to check each word in the list one by one.
    if len(words[i])>len(longest):       #This condition checks whether the current word is longer than the previously stored longest word.
        longest=words[i]                 # Update longest word

print("Longest word is:",longest)        
```

<h3>2) Write a program to check if a string contains only digits, only alphabets, or both.

```python
s = input("Enter a string: ")         

if s.isdigit():                        # Check if string has only digits
    print("String contains only digits")
elif s.isalpha():                      # Check if string has only alphabets
    print("String contains only alphabets")
elif s.isalnum():                      # Check if string has both letters and digits
    print("String contains both alphabets and digits")
else:
    print("String contains other characters")  # Spaces or special symbols
```
