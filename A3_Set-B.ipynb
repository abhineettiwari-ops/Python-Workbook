<h4>1) Write a program to find the second largest element in a tuple. 

```python
t = (5, 2, 8, 1, 9, 3)       

largest = t[0]               #Assume first element is the largest
second = t[0]                #Assume first element is the second largest
for i in range(len(t)):      
    if t[i] > largest:       #If current element is greater than largest
        second = largest     # Move largest value to second
        largest = t[i]       # Update largest value
    elif t[i] > second and t[i] != largest:  # Check for second largest
        second = t[i]        # Update second largest
print("Second largest element:", second)  
```

<h4>2) Write a program to sort a tuple of tuples based on the second element. 
<h4>Example: ((1, 5), (3, 2), (2, 8)) should become ((3, 2), (1, 5), (2, 8))

```python
t=((1,5),(3,2),(2,8))        
l=list(t)                    
for i in range(len(l)):      
    for j in range(i+1,len(l)):
        if l[i][1]>l[j][1]:  
            l[i],l[j]=l[j],l[i]  
t=tuple(l)                   
print(t)                    
```

<h4>3) Write a program to check if all elements in a tuple are unique. 

```python
t = (1, 2, 3, 4, 5)          
unique = True           #We first assume that all elements in the tuple are different.     
for i in range(len(t)):  #This loop checks every value in the tuple step by step.   
    if t.count(t[i]) > 1:    #This condition checks whether the current element occurs more than once in the tuple.
        unique = False       #This line tells that the tuple does NOT have all unique elements.
        break               
if unique:
    print("All elements are unique")
else:
    print("Tuple contains duplicate elements")
```

<h4>4) Write a program to swap the first and last element of a tuple.

```python
t=(10,20,30,40,50)         

first=t[0]                  #Store first element
last=t[-1]                  #Store last element
t=(last,)+t[1:-1]+(first,)  #Create new tuple with swapped values
print(t)                    
```
