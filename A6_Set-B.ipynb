<h4>1) Write a function to convert temperature from Celsius to Fahrenheit and vice versa.

<h5>:-Celsius to Fahrenheit

```python
def celsius_to_fahrenheit(c):          # Function is made to convert Celsius to Fahrenheit
    f=(c*9/5)+32                        # Formula is used to convert Celsius into Fahrenheit
    return f                            # Return converted temperature
```

<h5>:-Fahrenheit to Celsius

```python
def fahrenheit_to_celsius(f):          # Function is made to convert Fahrenheit to Celsius
    c=(f-32)*5/9                        # Formula is used to convert Fahrenheit into Celsius
    return c                   # Return converted temperature
```

<h5>:-Function Calls

```python
c=int(input("Enter temperature in Celsius: "))     # Take Celsius value from user
print("Fahrenheit:",celsius_to_fahrenheit(c))      # Call function and print result
f=int(input("Enter temperature in Fahrenheit: "))  # Take Fahrenheit value from user
print("Celsius:",fahrenheit_to_celsius(f))         # Call function and print result

```

<h4>2) Write a function that accepts a string and returns the count of vowels and consonants.

```python
def count_vowels_consonants(s):     
    vowels = "aeiouAEIOU"            # This string stores all vowel letters
    v_count = 0                      # Variable is used to count vowels
    c_count = 0                      # Variable is used to count consonants
    for ch in s:                   
        if ch.isalpha():             # Check only letters, ignore spaces and symbols
            if ch in vowels:         
                v_count = v_count+1  # Increase vowel count
            else:
                c_count = c_count+1  # Increase consonant count
    return v_count, c_count          # Return both counts
text = input("Enter a string: ")     # Take string input from user
v, c = count_vowels_consonants(text) # Call function and store returned values
print("Vowels:",v)
print("Consonants:",c)
```

<h4>3) Write a function that accepts variable number of arguments and returns their average.

```python
def find_average(*numbers):        # Function uses *numbers to accept many values
    total = 0                      # Variable is used to store sum of numbers
    for n in numbers:              # Loop is used to add each number one by one
        total = total + n          # Add current number to total
    avg = total / len(numbers)     # Divide total by count to get average
    return avg                     # Return the average value
result = find_average(10, 20, 30, 40)   # Function call with multiple values
print("Average is:", result)            # Print the average
```
