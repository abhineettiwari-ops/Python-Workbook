<h3>1) Write a Python Program to Calculate the Average of Numbers in a Given List.

```python
numbers = [20,25,50,60,15]
total = sum(numbers)          #to add all the numbers 
count = len(numbers)          #to find how many number are there 
average = total/count         #formula to find average Arithmetic Operator
print("The average of the number is:", average)
```

<h3>2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the 
values entered are duplicates otherwise it prints "ALL UNIQUE".

```python
numbers = []

for i in range(6):
    n = int(input("Enter a number: "))
    numbers.append(n)    # This adds the entered number 'n' to the list 'numbers'


unique = True   #assuming that all are unique

# Compare each number with every other number
for i in range(6):
    for j in range(i + 1, 6):       # Loop to compare the current number with the next numbers
        if numbers[i] == numbers[j]:   #If the number at position i is equal to the number at position j, then a duplicate exists."
            unique = False

# Print result
if unique:
    print("ALL UNIQUE")
else:
    print("DUPLICATES")
```

<h3>3) Write a program which accepts an integer value as input and print "Ok" if value is between 
1 to 50 (both inclusive) otherwise it prints "Out of range". 

```python
num = int(input("Enter a number: "))            #Taking input using input function


if num >= 1 and num <= 50:                      # Condition to check range if the number is between 1 and 50
    print("Ok")             
else:
    print("Out of range") 
```

<h3>4) Write a program which finds sum of digits of a number. 
Example: n=130 then output is 4 (1+3+0).

```python
num = int(input("Enter a number: "))
total = 0         # This variable stores the sum of the digits
while num > 0:    # This line tells Python to keep running the loop until the number becomes 0
     digit = num % 10    #Modulus operator (%)  used to get last digit 
     total = total + digit
     num = num // 10     # Remove last digit from number
print("Sum of digits is:", total)
 
```

<h3>5) Write a program which prints Fibonacci series of a number. 

```python
n = int(input("Enter number of terms: "))
# First two numbers of Fibonacci series are:-
a = 0
b = 1

for i in range(n):      # for loop to repeat n times
    print(a)            #this will print the current fibonaci number
    c = a + b           #this will add previous 2 numbers to get next number
    a = b               #These lines move the numbers forward so that the program can calculate the next Fibonacci number.
    b = c
```
