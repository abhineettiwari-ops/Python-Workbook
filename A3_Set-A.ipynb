<h4>1) Write a program to find the maximum and minimum element in a tuple without using 
max() and min() functions.

```python
t=(5,2,8,1,100,3)                 # Given tuple of numbers

max_val=t[0]                    # Assume first element is maximum
min_val=t[0]                    # Assume first element is minimum
for i in range(len(t)):         #This loop checks each element of the tuple one by one.
    if t[i]>max_val:           #Check if current element is greater than max
        max_val=t[i]            #Update maximum value
    if t[i]<min_val:             #Check if current element is smaller than min
        min_val=t[i]            #Update minimum value

print("Maximum element:",max_val)  
print("Minimum element:",min_val)  
```

<h4>2) Write a program to count the occurrence of each element in a tuple. 
<h4>Example: For tuple (1, 2, 3, 2, 1, 3, 2), output should show 1:2, 2:3, 3:2

```python
t=(1,2,3,2,1,3,2)              

for i in range(len(t)):        #This loop checks each element of the tuple one by one.
    if t.index(t[i])==i:       #This condition checks whether the current element is appearing for the first time in the tuple.
        print(t[i],":",t.count(t[i]))  #This statement prints the element and its total occurrence count in the tuple.
```

<h4>3) Write a program to convert a tuple of strings into a single string. 
<h4>Example: ("Hello", "World", "Python") should become "HelloWorldPython"

```python
t=("Hello","World","Python")      #Given tuple of strings
result=""                         #Empty string to store final result
for i in range(len(t)):           #This loop checks each value in the tuple one by one
    result=result+t[i]            #This statement concatenates each string element of the tuple into a single string.
print("Single string:",result)    
```

<h4>4) Write a program to find the sum of all even numbers in a tuple.

```python
t=(1,2,3,4,5,6,8)        # Given tuple
sum=0                   # Variable to store sum

for i in range(len(t)): # Loop through tuple
    if t[i]%2==0:       # Check if number is even
        sum=sum+t[i]    # Add even number

print("Sum of even numbers:",sum)
```

<h4>5) Write a program to remove duplicates from a tuple (return a new tuple with unique 
elements).

```python
t = (1, 2, 3, 2, 4, 1, 5)      
unique = ()                   # Empty tuple to store unique elements

for i in range(len(t)):       # Loop through the tuple
    if t[i] not in unique:    ## Check if the element is not already present in the unique tuple
        unique = unique + (t[i],)  ## This line adds the new element to the unique tuple.

print("Tuple without duplicates:", unique)  
```
